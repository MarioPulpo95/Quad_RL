```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import rospkg
from mpl_toolkits.mplot3d import Axes3D
```

```python
rospack = rospkg.RosPack()
pkg_path = rospack.get_path('rotors_gazebo')
train_path = pkg_path + '/Training/Tensorboard/'
df_sac_train = pd.read_csv(train_path + 'run-NewTrains_SAC4_NOTILT_1-tag-rollout_ep_rew_mean.csv')
df_ddpg_gaussian_train = pd.read_csv(train_path + 'run-NewTrains_DDPG0_NOTILT_GAUSSIAN_1-tag-rollout_ep_rew_mean.csv')
df_ddpg_ou_train = pd.read_csv(train_path + 'run-NewTrains_DDPG0_NOTILT_ORNSTEIN_1-tag-rollout_ep_rew_mean.csv')
df_td3_train = pd.read_csv(train_path + 'run-NewTrains_TD30_NOTILT_1-tag-rollout_ep_rew_mean.csv')
df_ppo_train = pd.read_csv(train_path + 'run-NewTrains_PPO0_NOTILT_1-tag-rollout_ep_rew_mean.csv')

df_ddpg_gauss_episode = pd.read_csv(train_path + 'run-NewTrains_DDPG0_NOTILT_GAUSSIAN_1-tag-rollout_ep_len_mean.csv')
df_ddpg_ou_episode = pd.read_csv(train_path + 'run-NewTrains_DDPG0_NOTILT_ORNSTEIN_1-tag-rollout_ep_len_mean.csv')
df_td3_episode = pd.read_csv(train_path + 'run-NewTrains_TD30_NOTILT_1-tag-rollout_ep_len_mean.csv')
df_ppo_episode = pd.read_csv(train_path + 'run-NewTrains_PPO0_NOTILT_1-tag-rollout_ep_len_mean.csv')
df_sac_episode = pd.read_csv(train_path + 'run-NewTrains_SAC4_NOTILT_1-tag-rollout_ep_len_mean.csv')
```

```python
y_sac = df_sac_train['Value'].values
x_sac = df_sac_train['Step'].values

y_td3 = df_td3_train['Value'].values
x_td3 = df_td3_train['Step'].values

y_ddpg_gaussian = df_ddpg_gaussian_train['Value'].values
x_ddpg_gaussian = df_ddpg_gaussian_train['Step'].values

y_ddpg_ou = df_ddpg_ou_train['Value'].values
x_ddpg_ou = df_ddpg_ou_train['Step'].values
y_ppo = df_ppo_train['Value'].values
x_ppo = df_ppo_train['Step'].values

y_ddpg_gauss_ep = df_ddpg_gauss_episode['Value'].values
x_ddpg_gauss_ep = df_ddpg_gauss_episode['Step'].values

y_ddpg_ou_ep = df_ddpg_ou_episode['Value'].values
x_ddpg_ou_ep = df_ddpg_ou_episode['Step'].values

y_td3_ep = df_td3_episode['Value'].values
x_td3_ep = df_td3_episode['Step'].values

y_ppo_ep = df_ppo_episode['Value'].values
x_ppo_ep = df_ppo_episode['Step'].values

y_sac_ep = df_sac_episode['Value'].values
x_sac_ep = df_sac_episode['Step'].values
plt.figure(figsize=(15,7))

plt.plot(x_ddpg_gaussian, y_ddpg_gaussian) 
plt.plot(x_ddpg_ou, y_ddpg_ou) 
plt.plot(x_td3, y_td3)  
plt.plot(x_sac, y_sac)
plt.plot(x_ppo, y_ppo)
plt.xlabel('Steps')
plt.ylabel('Ricompense')
plt.legend(['DDPG GAUSSIAN', 'DDPG OU', 'TD3', 'SAC','PPO'],loc='best')
plt.title('Learning Curves')

print(np.mean(y_ddpg_gaussian))
print(np.mean(y_ddpg_ou))
print(np.mean(y_td3))
print(np.mean(y_sac))
print(np.mean(y_ppo))

plt.figure(figsize=(15,7))
plt.plot(x_ddpg_gauss_ep, y_ddpg_gauss_ep) 
plt.plot(x_ddpg_ou_ep, y_ddpg_ou_ep) 
plt.plot(x_td3_ep, y_td3_ep) 
plt.plot(x_sac_ep, y_sac_ep)
plt.plot(x_ppo_ep, y_ppo_ep) 
plt.xlabel('Steps')
plt.ylabel('Duration')
plt.legend(['DDPG ORN','DDPG OU', 'TD3','SAC','PPO'],loc='best')
plt.title('Episodes Duration ')
plt.show()

print(np.mean(y_ddpg_gauss_ep))
print(np.mean(y_ddpg_ou_ep))
print(np.mean(y_td3_ep))
print(np.mean(y_sac_ep))
print(np.mean(y_ppo_ep))

```
